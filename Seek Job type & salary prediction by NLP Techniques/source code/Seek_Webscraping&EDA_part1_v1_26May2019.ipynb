# Scraping to get all from pages

```python
#R_urls_text_list

R_All = []
R_Job_list = []


#def mini_url_loop:
for i,item in enumerate(R_urls_text_list):
    #print(i)
    R_url_response = requests.get(R_urls_text_list[i])
    R_url_list_1_html = R_url_response.text    
    Company_name = Selector(text= R_url_list_1_html).xpath('//*[@id="app"]//div/h2/span[2]/span/text()').extract()
    Job_date = Selector(text= R_url_list_1_html).xpath('//*[@id="app"]//div[2]/div/article/div[1]/div[3]/div/section/dl/dd[1]/span/span/text()').extract()
    R_Salary = Selector(text= R_url_list_1_html).xpath('//*[@id="app"]/div/div[3]/div/div/div/div[2]/div/article/div[1]/div[2]/div/section//div[1]/dd/span/span/text()').extract()
    Description = Selector(text= R_url_list_1_html).xpath('//*[@id="app"]//*[@class="_2e4Pi2B"]').extract()
    R_Job_title = Selector(text= R_url_list_1_html).xpath('//*[@id="app"]/div/div[3]/div/div/div/div[2]/div/article/div[1]/div[1]/section/div/span/span/h1/text()').extract() 
    #company_name_list.append(company_name)
    #Job_date_list.append(Job_date)
    #Salary_list.append(Salary)
    #Description_list.append(Description)
    R_All = [R_urls_text_list[i],Company_name,R_Job_title, Job_date, R_Salary, Description]
    R_Job_list.append(R_All)
    
R_Job_list    
```
