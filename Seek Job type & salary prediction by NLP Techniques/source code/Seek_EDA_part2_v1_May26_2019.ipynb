```python
def job_title_map(title_string):
    title_string =  title_string.lower()
    clean_analyst = re.findall('analyst',title_string)
    clean_scientist = re.findall('scientist',title_string)
    clean_manager = re.findall('manager',title_string)
    
    if clean_analyst == ['analyst']:    
            return('Data Analyst')
    elif clean_scientist == ['scientist']:  
            return('Data Scientist')
    elif clean_manager == ['manager']:
            return('Data manager')
    else:                
            print('no info')
            
        
```

```python
Entry_level_list = ['entry','trainee','intern','graduate']
Junior_level_list = ['junior']
Mid_level_list = ['mid']
Senior_level_list = ['senior','lead']
Manager_level_list = ['Manager','manager']


def job_level_map(level_string):
    level_string =  level_string.lower()
    clean_Trainee = re.findall('trainee',level_string)
    clean_intern = re.findall('intern',level_string) 
    clean_graduate = re.findall('graduate',level_string)
    clean_entry = re.findall('entry',level_string)                                     
    clean_Junior = re.findall('junior',level_string)
    clean_mid = re.findall('mid',level_string)
    clean_senior = re.findall('senior',level_string)
    clean_lead = re.findall('lead',level_string)
    clean_manager = re.findall('manager',level_string)
        
    if ((clean_Trainee == ['trainee']) or   (clean_intern ==  ['intern']) or (clean_graduate ==  ['graduate'])
        or (clean_entry == ['entry'])):        
            return('Trainee')
    elif clean_Junior == ['junior']:  
            return('Junior')
    elif clean_mid == ['mid']:  
            return('mid')   
    elif clean_senior == ['senior']:  
            return('senior')         
    elif clean_lead == ['lead']:  
            return('lead')              
    elif clean_manager == ['manager']:
            return('manager')
    else:                
            print('no level')
 
```
